###  Modified calculate_total_price Functions 

````python
def calculate_total_price(products):
    """
    Prompts the user for the price of each product and calculates the total inventory value.
    Includes error handling for invalid (non-numeric or negative) price inputs.
    """
    total_value = 0
    product_prices = {}

    for product in products:
        valid_price = False
        while not valid_price:
            try:
                # Prompt and attempt to convert input to a float
                price = float(input(f"Enter the price of {product}: "))

                # Check for negative value (custom validation)
                if price < 0:
                    # Raise a ValueError to be caught by the except block
                    raise ValueError("Price cannot be negative. Please enter a non-negative value.")

                # If successful, store the price and exit the loop
                product_prices[product] = price
                total_value += price
                valid_price = True

            except ValueError as e:
                # Handles non-numeric input (from float()) AND the custom negative value error
                print(f"Error: {e}. Please enter a valid non-negative number for the price.")

    # Note: If the actual goal is to calculate inventory value (price * quantity), 
    # the inventory dictionary would also be needed here. 
    # Based on the original prompt, this version just gets the price for each product.
    print(f"\nTotal value (sum of all product prices entered): ${total_value:.2f}")
    return product_prices
````

### Modified get_customer_orders Function

````python
def get_customer_orders(inventory):
    """
    Prompts the user for customer orders, validating the number of orders and product names/stock.
    """
    customer_orders = {}

    # --- Error Handling for Number of Orders ---
    num_orders = -1
    while num_orders < 0:
        try:
            num_orders = int(input("\nEnter the total number of customer orders to process: "))
            if num_orders < 0:
                print("Error: The number of orders cannot be negative. Please try again.")
                num_orders = -1 # Reset to keep the loop going
        except ValueError:
            print("Error: Invalid input. Please enter a whole number for the number of orders.")

    # --- Getting Product Orders with Validation ---
    print(f"Processing {num_orders} orders...")

    for i in range(num_orders):
        valid_order = False
        while not valid_order:
            try:
                # Prompt for product name
                product_name = input(f"Enter product name for Order #{i + 1}: ").strip().capitalize()

                # 1. Check if the product exists in inventory
                if product_name not in inventory:
                    # Use a custom exception or raise a standard one for clarity
                    raise KeyError(f"Product '{product_name}' not found in inventory.")

                # 2. Check if the product is in stock
                if inventory[product_name] <= 0:
                    raise ValueError(f"Product '{product_name}' is currently out of stock.")

                # If successful, record the order
                customer_orders[product_name] = customer_orders.get(product_name, 0) + 1
                inventory[product_name] -= 1 # Decrease stock for simplicity
                valid_order = True

            except KeyError as e:
                # Handles product not found in inventory
                print(f"Error: {e}. Please enter a product name that is in stock.")
            except ValueError as e:
                # Handles product found but out of stock
                print(f"Error: {e}. Please choose another product.")
            except Exception:
                # Catch-all for unexpected errors
                print("An unexpected error occurred. Please try again.")

    return customer_orders
````
